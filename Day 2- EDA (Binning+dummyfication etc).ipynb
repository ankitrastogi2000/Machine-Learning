# 5
1.Calculate outliers for a given DataCancer dataset .

2.First calculate first and third quartile (Q1)(25%) , Q3(75%). Find IQR(inter quartile range) = Q3-Q1

Finally compute Q1 - 1.5IQR and Q3 + 1.5IQR for feature radius mean columns  3. You can see outliers also from box plots . plot box plots for the feature radius mean column.

```python
import pandas as pd
gr4 = pd.read_csv('DataCancer.csv')
```

```python
df4 = pd.DataFrame(gr4)
```

```python
df4.head()
```

```python
import numpy as np
outliers=[]
def detect_outlier(data):
    
    threshold=3
    mean_1 = np.mean(data)
    std_1 =np.std(data)
    
    
    for y in data:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers
for column in df4.drop(['diagnosis'], axis=1):
    outlier_datapoints = detect_outlier(df4[column])
    
```

```python
print(outlier_datapoints)
```

```python
q1, q3= np.percentile(df4['radius_mean'],[25,75])
iqr = q3 - q1
```

```python
lower_bound = q1 -(1.5 * iqr)
print(lower_bound)
```

```python
upper_bound = q3 +(1.5 * iqr)
print(upper_bound)
```

```python
import matplotlib.pyplot as plt

figure = plt.figure(figsize = (10, 7))
plt.boxplot(df4['radius_mean'])
plt.show()
```

# 10.
Normal(Gaussian) Distribution
We have data of students with an average IQ of 110. There are few people that are super intelligent and
their IQs are higher than 110. It can be 140 or 150 but it is rare. Also there are few people that are low
intelligent and their IQ is lower than 110. It can be 40 or 50 but it is rare. From this information we can
say that the mean of IQ is 110. And lets say standard deviation is 20. Mean and standard deviation is
parameters of normal distribution. Lets create a 100000 sample and visualize it with a histogram.Z-Score
What if I want to know what percentage of people should have an IQ score between 80 and 140? (use
z-score to answer this question.)
* z = (x - mean)/std

```python
import matplotlib.pyplot as plt
iq = np.random.normal(110, 20, 100000)
figure = plt                                                                                   #.figure(figsize = (10, 10))
plt.hist(iq)
plt.show()
```

```python
iq.mean()
```

```python
z_score = []
for i in iq:
    if i >= 80 and i <= 140:
        z_score.append((i-110)/20)
figure = plt                                                                                                                     #.figure(figsize = (10, 10))
plt.hist(z_score)
plt.show()
```

```python
s=0
for i in iq:
    if i >= 80 and i <= 140:
        s=s+i
s1=iq.sum()
print("Percentage of people who should have an IQ score between 80 and 140 is:\n",(s/s1)*100)
```
